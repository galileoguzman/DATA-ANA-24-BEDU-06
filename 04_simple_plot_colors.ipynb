```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Datos que importar o trabajar

categorias = ['A', 'B', 'C', 'D', 'E'] 
valores = [5, 8, 3, 3, 2]

# Figura de grafica de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Colors
sns.barplot(
    x=categorias,
    y= valores,
    ax=ax,
    # Esto cambiamos
    palette=sns.color_palette("Oranges", n_colors=len(categorias)),
    hue=categorias,
    legend=False
)


plt.show()
```
